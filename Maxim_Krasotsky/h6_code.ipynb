```python
# Task 4.1
class Counter:
    def __init__(self, start=0, stop=-1):
        self.start = start
        self.counter = self.start
        self.stop = stop
        
    def increment(self):
        try:
            temp = self.counter + 1
            if temp == self.stop+1:
                raise ValueError
            self.counter = temp
        except ValueError:
            print("Maximal value is reached.")
    
    def get(self):
        print(self.counter)
    
    
c1 = Counter(start=42)
c1.increment()
c1.get()

c2 = Counter()
c2.increment()
c2.get()
c2.increment()
c2.get()

c3 = Counter(start=42, stop=43)
c3.increment()
c3.get()
c3.increment()
c3.get()
```

```python
# Task 4.3
import string


class Cipher:
    __alphabet = string.ascii_uppercase

    @staticmethod
    def __make_code_word(word):
        code_word = ""
        for ch in word:
            if ch not in code_word:
                code_word += ch
        return code_word

    @staticmethod
    def __make_code_alphabet(code_word):
        code_alphabet = string.ascii_uppercase
        for ch in code_word:
            code_alphabet = code_alphabet.replace(ch, "")
        code_alphabet = code_word + code_alphabet
        return code_alphabet

    def __init__(self, word):
        self.code_word = self.__make_code_word(word).upper()
        self.__code_alphabet = self.__make_code_alphabet(self.code_word)

    def encode(self, string):
        encode_string = ""
        for ch in string:
            if ch == " ":
                encode_string += " "
            elif ch.isupper():
                encode_string += self.__code_alphabet[Cipher.__alphabet.index(ch)]
            else:
                encode_string += self.__code_alphabet[Cipher.__alphabet.index(ch.upper())].lower()
        print(encode_string)

    def decode(self, string):
        decode_string = ""
        for ch in string:
            if ch == " ":
                decode_string += " "
            elif ch.isupper():
                decode_string += Cipher.__alphabet[self.__code_alphabet.index(ch)]
            else:
                decode_string += Cipher.__alphabet[self.__code_alphabet.index(ch.upper())].lower()
        print(decode_string)


cipher = Cipher("cryptocr")
cipher.encode("Hello world")
cipher.decode("Fjedhc dn atidsn")
```

```python
# Task 4.4
class Bird:
    def __init__(self, name):
        self.name = name
        
    def fly(self):
        print(f"{self.name} can fly")
    
    def walk(self):
        print(f"{self.name} can walk")
        
    def __str__(self):
        return f"{self.name} can walk and fly"
    
    
class FlyingBird(Bird):
    def __init__(self, name, ration="grains"):
        super().__init__(name)
        self.ration = ration
        
    def eat(self):
        print(f"It eats mostly {self.ration}")
        

class NonFlyingBird(Bird):
    def __init__(self, name, ration="fish"):
        super().__init__(name)
        self.ration = ration
        
    def eat(self):
        print(f"It eats mostly {self.ration}")
        
    def fly(self):
        try:
            raise AttributeError
        except AttributeError:
            print(f"{self.name} object has no attribute 'fly'")
        
    def swim(self):
        print(f"{self.name} bird can swim")
        
    def __str__(self):
        return f"{self.name} can walk and swim"
        

class SuperBird(NonFlyingBird, FlyingBird):
    def __init__(self, name):
        super().__init__(name)
    
    def __str__(self):
        return f"{self.name} can walk, swim and fly"
    
    

b = Bird("Any")
b.walk() 
print()
p = NonFlyingBird("Penguin", "fish")
p.swim()
p.fly()
p.eat()
print()
c = FlyingBird("Canary")
print(str(c))
c.eat()
print()
s = SuperBird("Gull")
print(str(s))
s.eat()
```

```python
# Task 4.7
class Money:
    exchange_rate = {
        'EUR': 0.93,
        'BYN': 2.5,
        'RUB': 72,
        'USD': 1
    }
    
    def __init__(self, value: float, exchange: str = "USD"):
        self.value = value
        self.exchange = exchange
        
    def __str__(self):
        return f"{self.value:.2f} {self.exchange}"
        
    def translate_to_ye(self) -> float:
        return self.value / Money.exchange_rate[self.exchange]
    
    def __add__(self, other):
        return Money(self.translate_to_ye() + other.translate_to_ye())
    
    def __sub__(self, other):
        return Money(self.translate_to_ye() - other.translate_to_ye())
    
    def __mul__(self, other):
        return Money(self.translate_to_ye() * other.translate_to_ye())
    
    def __truediv__(self, other):
        return Money(self.translate_to_ye() / other.translate_to_ye())
        
    def __eq__(self, other):
        return self.translate_to_ye() == other.translate_to_ye()
    
    def __ne__(self, other):
        return self.translate_to_ye() != other.translate_to_ye()
    
    def __lt__(self, other):
        return self.translate_to_ye() < other.translate_to_ye()
    
    def __gt__(self, other):
        return self.translate_to_ye() > other.translate_to_ye()
    
    def __le__(self, other):
        return self.translate_to_ye() <= other.translate_to_ye()
    
    def __ge__(self, other):
        return self.translate_to_ye() >= other.translate_to_ye()
    
    
import operator
  
    
sign_list = ['+', '-', '*', '/', '==', '!=', '>', '<', '>=', '<=']
    
def print_operation(value_1, value_2, index):
    print(value_1, sign_list[index], value_2, ' -> ', end=" ")
    
    
u1 = Money(1.86, "EUR")
u2 = Money(72, "RUB")
u3 = Money(100.0)
u4 = Money(200.0)

print_operation(u1, u2, 0)
print(u1 + u2)
print_operation(u3, u4, 0)
print(u3 + u4)
print()
print_operation(u1, u2, 1)
print(u1 - u2)
print_operation(u3, u4, 1)
print(u3 - u4)
print()
print_operation(u1, u2, 2)
print(u1 * u2)
print_operation(u3, u4, 2)
print(u3 * u4)
print()
print_operation(u1, u2, 3)
print(u1 / u2)
print_operation(u3, u4, 3)
print(u3 / u4)
print()
print_operation(u1, u2, 4)
print(u1 == u2)
print_operation(u3, u4, 4)
print(u3 == u4)
print()
print_operation(u1, u2, 5)
print(u1 != u2)
print_operation(u3, u4, 5)
print(u3 != u4)
print()
print_operation(u1, u2, 6)
print(u1 > u2)
print_operation(u3, u4, 6)
print(u3 > u4)
print()
print_operation(u1, u2, 7)
print(u1 < u2)
print_operation(u3, u4, 7)
print(u3 < u4)
print()
print_operation(u1, u2, 8)
print(u1 >= u2)
print_operation(u3, u4, 8)
print(u3 >= u4)
print()
print_operation(u1, u2, 9)
print(u1 <= u2)
print_operation(u3, u4, 9)
print(u3 <= u4)
```

```python
# Task 4.8
from math import ceil


class Pagination:
    def __init__(self, text: str, text_in_page):
        self.text = text
        self.text_in_page = text_in_page
        self.item_count = len(text)
        self.page_count = ceil(self.item_count / text_in_page)
        self.pages = self.__slice_into_pages(text_in_page)
            
    def __slice_into_pages(self, text_in_page):
        start = 0
        end = text_in_page
        pages = []
        for i in range(self.page_count):
            pages.append(self.text[start:end])
            start = end
            end += text_in_page
        return pages
    
    def count_items_on_page(self, page_number):
        try:
            return len(self.pages[page_number])
        except IndexError:
            print("Exception: Invalid index. Page is missing")
            
    def display_page(self, page_number):
        print(self.pages[page_number])
        
    def find_page(self, msg):
        if msg not in self.text:
            raise Exception (f"'{msg}' is missing on the pages")
        find_pages = set()
        start_indexes = [i for i in range(len(self.text)) if self.text.startswith(msg, i)]
        for index in start_indexes:
            find_pages.add(index // self.text_in_page)
            find_pages.add((index + len(msg)) // self.text_in_page)
        return list(find_pages)
        

pages = Pagination('Your beautiful text', 5)
print(pages.page_count)
print(pages.item_count)
print(pages.count_items_on_page(0))
print(pages.count_items_on_page(3))
#print(pages.count_items_on_page(4))
pages.display_page(0)
print(pages.find_page('e'))
print(pages.find_page('beautiful'))
print(pages.find_page('great'))
```
